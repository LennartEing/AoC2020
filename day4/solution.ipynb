```python
import re

needed_values = [
    "byr",
    "iyr",
    "eyr",
    "hgt",
    "hcl",
    "ecl",
    "pid",
]

filename = "./input.txt"
with open(filename) as f:
    content = f.readlines()
content = [line.strip() for line in content]
idx_list = [idx + 1 for idx, val in enumerate(content) if val == ''] 
content = [content[i: j] for i, j in zip([0] + idx_list, idx_list + ([len(content)] if idx_list[-1] != len(content) else []))]
content = [list(filter(('').__ne__, line)) for line in content]
passports = [' '.join(passport) for passport in content]

def check_validity(passport, values):
    return all([not re.search(value, passport) is None for value in values])
    

valid_passports = [passport for passport in passports if check_validity(passport, needed_values)]
print(len(valid_passports))
```

```python
classes = [
    "byr:(?P<year>[0-9]{4})",
    "iyr:(?P<year>[0-9]{4})",
    "eyr:(?P<year>[0-9]{4})",
    "hgt:(?P<hgt>(([5-7][0-9]in)|(1[5-9][0-9]cm)))"
    "ecl:(?P<ecl>(amb|blu|brn|gry|grn|hzl|oth))"
    "hcl:(?P<hcl>#([0-9]|[a-f]){6})"
    "pid:(?P<pid>[0-9]{9})"
]

for passport in passports:
    for clazz in classes:
        print(re.match(clazz, passport))
        
```
